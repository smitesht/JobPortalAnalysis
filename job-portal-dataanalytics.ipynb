### Clean and make Payrate feature more meaningful

```python
# Observe payrate column

df['payrate'].head(20)
```

```python
'''From the above payrate feature values we can observe that payrate in a range (Minimum Payrate - Max Payrate).
However, all the values do not contains range and contains other values such as "Not Disclosed by Recruiter", 1,00,000 - 2,00,000 P.A. incentives for achiev
We need to clean the feature and make it more useful. 
Our goal is to clean the payrate feature and construct min_pay and max_pay features from payrate
'''

# First analyze how many valid range values. Make those values as a min_pay and max_pay and others NaN
# The value like 1,50,000 - 2,25,000 P.A is a valid value others are invalid

payrate_len = []

for pay in df['payrate']:
    payrate_len.append(len(str(pay).split('-')))    

pd.Series(payrate_len).value_counts()

# from the result we can observe that there are 4682 values has length 2 which are valid entries while rest of the entries are invalid 
```

#### Extract min_payrate and max_payrate

```python
# Steps to extract min_pay and max_pay
# 1. Split the payrate by '-'
# 2. perform certian operation and extract min_pay and max_pay

payrate_split = df['payrate'].str.split('-',expand=True)
payrate_split
```

```python
# col 0 and 1 contain the min_pay and max_pay

# trim data of col 0
payrate_split[0] = payrate_split[0].str.strip()
# remove , from the values
payrate_split[0] = payrate_split[0].str.replace(',','')
payrate_split[0]
```

```python
# Now we need to clean col 1. We can observe that col 1 has value ' 2,25,000 P.A'
# We can use Regex for cleaning, define a pattern of digits only

# Trim data
payrate_split[1] = payrate_split[1].str.strip()
# Replace , with black
payrate_split[1] = payrate_split[1].str.replace(',','')

pattern = r'\D.*'
payrate_split[1] = payrate_split[1].str.replace(pattern,'',regex=True)
payrate_split[1]
```

```python
# Convert to numerical. If the data is string, it convert into NaN
payrate_split[0] = pd.to_numeric(payrate_split[0], errors='coerce')
payrate_split[1] = pd.to_numeric(payrate_split[1], errors='coerce')
```

```python
payrate_split[1]
```

```python
payrate_split.dtypes
```

```python
# append new min_pay (payrate_split[0]) and max_pay (payrate_split[1]) to main dataframe df

df['min_pay'] = payrate_split[0]
df['max_pay'] = payrate_split[1]

df.head()
```

```python
# Create average payrate feature from min_pay and max_pay.

df['avg_pay'] = (payrate_split[0] + payrate_split[1]) / 2
```

```python
# Create Bar chart

def showval(pct):
    return "{:.1f}".format(pct)


plt.figure(figsize=(12,6))
plt.pie(labels=company_df['company'][0:10],x= company_df['Number of jobs'][0:10],autopct= lambda pct: showval(pct) )
```

```python
# We can make single method that show bar chart / pie chart of any categorical columns 

def perform_analysis(data, feature, col1, col2, chart_type, chart_title="",color="#05c46b"):
    
    '''
    This function display either Bar chart or Pie chart based on given data.
    
    data: dataframe
    
    feature: column_name that we have to consider for analysis
    
    col1: 1st column name that you want to assign to new df
    
    col2: 2nd column name that you want to assign to new df
    
    @chart_type: if its value is 1, display Bar chart
    @chart_type: if its value is 2, display Pie chart
    
    chart_title: This is a char title
    
    color: This present Bar color in Hex
    '''
    
    dataframe = data[feature].value_counts().reset_index()
    dataframe.columns = [col1,col2]
    
    print(dataframe.head(5))
    
    plt.figure(figsize=(12,6))    
    
    if chart_type == 1:        
        plt.bar(dataframe[col1][0:10], dataframe[col2][0:10], color=color)
        plt.xticks(rotation='vertical')        
        #plt.title(label='Number of Users of a particular Language',
        #  fontweight=10,
        #  pad='2.0')
        
    elif chart_type == 2:
        def showval(pct):
            return "{:.1f}".format(pct)        
        
        plt.pie(labels=dataframe[col1][0:10],x= dataframe[col2][0:10],autopct= lambda pct: showval(pct) )
    
```

```python
perform_analysis(new_df,'company','company','Number of jobs',1,"Top 10 Companies by Jobs Posting")
```
